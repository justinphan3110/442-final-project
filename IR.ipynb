```python
import pickle
from hal import HAL
import numpy as np
import pandas as pd
import os
```

```python
def addTsne(data):
        
        dirc = os.getcwd()
        for file in os.listdir(dirc):
            if file.find('info_hal') > -1:
                for file in os.listdir(str(dirc) + '/info_hal'):
                    if file.find('tsne') > -1:
                        tsne = pd.read_pickle(str(dirc) + '/info_hal/' + file)
                        break
        col1 = []
        col2 = []
        for x in tsne:
            col1.append(x[0])
            col2.append(x[1])
        data["TSNE1"] = col1
        data["TSNE2"] = col2

        return data
```

```python
data = pickle.load(open('Lupus/final_lupus_data.pkl','rb'))

data = data.drop(['Label_0.9'],1)

model = HAL(clf_type='svm', outlier_ratio=0.1, late_exag=900, alpha_late=2.0, n_cluster_init=150,warm_start=True)

train_data = data.drop(['score', ' sample type'],1).values[np.random.choice(list(range(len(data))),100000)]

model.fit(train_data)

labels_tSNE = model.predict(train_data, cv=0.8)
```

```python
output = pd.DataFrame()
output["clusters"] = labels_tSNE
output = addTsne(output)
!mkdir out
pickle.dump(output, open("out/tsne_data"  + "_" +str(0.8) + ".pkl", "wb"))


```

```python
def split_dataframe(df, chunk_size = 1000000): 
    chunks = list()
    num_chunks = len(df) // chunk_size + 1
    for i in range(num_chunks):
        chunks.append(df[i*chunk_size:(i+1)*chunk_size])
    return chunks
chunks = split_dataframe(data.drop(['score', ' sample type'],1))
```

```python
# data['clusters'] = model.predict(data.drop(['score', ' sample type'],1).values,cv=0.8)
for index, chunk in enumerate(chunks):
  chunk['clusters'] = model.predict(chunk.values,cv=0.8)
  chunks[index] = chunk
```
